```
import numpy as np
import pandas
from matplotlib import pyplot as plt
from numpy import random



# Defining hypothesis and cost function
def hy(x1, x2, w):
    p = []
    for i in range(len(x1)):
        p.append((w[0] + w[1] * x1[i] + w[2] * x2[i]))
    return np.array(p)


def CF(h, y):
    sum = 0
    for i in range(len(y)):
        sum = sum + (h[i] - y[i]) ** 2
    return sum / 2


# Importing data files
a = [1, 2]
file1 = pandas.read_excel("training_feature_matrix.xlsx", header=None, names=a)
xtr1 = np.array(file1[1].values.tolist())
xtr2 = np.array(file1[2].values.tolist())

file2 = pandas.read_excel("training_output.xlsx", header=None, names=a)
ytr = np.array(file2[1].values.tolist())

file3 = pandas.read_excel("test_feature_matrix.xlsx", header=None, names=a)
xte1 = np.array(file3[1].values.tolist())
xte2 = np.array(file3[2].values.tolist())

file4 = pandas.read_excel("test_output.xlsx", header=None, names=a)
yte = np.array(file4[1].values.tolist())

# Normalising Training Data and test data

Xtr1 = (xtr1 - np.mean(xtr1)) / np.std(xtr1)
Xtr2 = (xtr2 - np.mean(xtr2)) / np.std(xtr2)
Y = (ytr - np.mean(ytr)) / np.std(ytr)
Xte1 = (xte1 - np.mean(xtr1)) / np.std(xtr1)
Xte2 = (xte2 - np.mean(xtr2)) / np.std(xtr2)

# initialising Weight values
w = np.array([0, 0, 0])

# giving Alpha and no. of iterations
alpha = 0.0001
t = 600
lam = 0.00001

h = hy(Xtr1, Xtr2, w)

# Creating a array for storing cost function values
jw = np.empty(shape=(t + 1, 1))
jw[0] = (CF(h, Y))

# Creating a array for storing weight values
W = np.empty(shape=(t + 1, 3))
W[0] = w



# iteration Loop
for i in range(1, t + 1):
    # here s s1 s2 are the sigma part in weight update rule
    s = h - Y
    random.shuffle(s)
    W[i, 0] = W[i - 1, 0] - (alpha * (s.sum()))
    s1 = (h - Y) * Xtr1
    random.shuffle(s1)
    W[i, 1] = W[i - 1, 1] - (alpha * (s1.sum()))
    s2 = (h - Y) * Xtr2
    random.shuffle(s2)
    W[i, 2] = W[i - 1, 2] - (alpha * (s2.sum()))
    h = hy(Xtr1, Xtr2, W[i])
    jw[i] = (CF(h, Y))

# calculating instance value of the minimum cost function
instance = np.where(jw == np.amin(jw))
hte = hy(Xte1, Xte2, np.array(W[instance[0][0]]))
Hte = (hte * np.std(ytr)) + np.mean(ytr)

# calculating mean square error between Denormalized hypothesis and test output
mse = (CF(Hte, yte)* 2) / len(Xte1)
print("mean square error between Denormalized hypothesis and test output:")
print(mse)
print(W[instance[0][0]])
print(min(jw))

# creating an array for iterations
it = []
for i in range(len(jw)):
    it.append(i + 1)

# plot between cost function and iterations
plt.figure(1)
plt.plot(np.array(it), jw)
plt.title('Cost Function Vs Iterations : Batch Gradient Descent')
plt.xlabel('Iterations')
plt.ylabel('Cost Function')
plt.figure(2)
ax=plt.axes(projection="3d")
ax.plot3D((W[0:,1]),(W[0:,2]),(jw.reshape(1,len(jw))[0]))
plt.show()

```
